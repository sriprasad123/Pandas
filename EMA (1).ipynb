```python
# Columns to search through
COLS = ['Medicine name', 
        'Therapeutic area',
       'International non-proprietary name (INN) / common name',
       'Active substance']
```

```python
def filter_one_column(df: 'pd.DataFrame', column: 'str', text: 'str') -> 'pd.DataFrame':
    """Return a subset of dataframe whose column contains matches (regex-based) of specified text"""
    return df[df[column].str.contains(text, na=False, case=False, regex=True)] 
```

```python
def filter_columns(df: 'pd.DataFrame', columns: "List[Str]", text: 'str') -> 'pd.DataFrame':
    """Return a subset of dataframe whose columns contan matches (regex-based) of specified text"""
    list_out = []
    for col in columns:
        list_out.append(filter_one_column(df, col, text))
    df_out = pd.concat(list_out)
    df_out = df_out.drop_duplicates()
    return df_out
```
